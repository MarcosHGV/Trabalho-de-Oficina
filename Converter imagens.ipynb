```python
from mtcnn import MTCNN
import cv2
import os
```

```python
detector = MTCNN()
```

```python
# Treino
img_dir = r"D:\UTF\7 Periodo\Oficina\DataSet\Teste\Marcos"
save_dir = r"D:\UTF\7 Periodo\Oficina\DataSet\Teste\Marcos"
```

```python
for img_name in os.listdir(img_dir):
    if img_name.endswith(".jpg") or img_name.endswith(".png"):  # extensÃµes de imagem
        img_path = os.path.join(img_dir, img_name)
        img = cv2.imread(img_path)
        height, width = img.shape[:2]
        
        # Detectar faces
        faces = detector.detect_faces(img)
        
        # Criar o arquivo .txt correspondente
        with open(os.path.join(save_dir, f"{img_name.split('.')[0]}.txt"), 'w') as f:
            for face in faces:
                # Converter bounding box para o formato YOLO
                x, y, w, h = face['box']
                x_center = (x + w / 2) / width
                y_center = (y + h / 2) / height
                w_norm = w / width
                h_norm = h / height
                
                # Escrever as coordenadas da caixa delimitadora no formato YOLO
                f.write(f"0 {x_center} {y_center} {w_norm} {h_norm}\n")
```
